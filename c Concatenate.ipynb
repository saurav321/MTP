```python
import os
import glob
import pandas as pd

def concatenate(indir=f"/home/sauravadhikari/MTP/data/data_cleaned/", outfile=f"/home/sauravadhikari/MTP/data/Master_Final.csv"):
    os.chdir(indir)
    fileList=glob.glob("*.csv")
    dfList=[]
    #colnames=["Origin","Destination","Dept_Date","Dept_Time","Dept_Weekday","Total_Fare","Airways","Class_Type","Flight Number","Flight Code","Taken","Taken_Day"]
    for filename in fileList:
        print(filename)
        df=pd.read_csv(filename, header=0)
        dfList.append(df)
    concatDf=pd.concat(dfList,axis=0, ignore_index=True)
    #concatDf.columns=colnames
    concatDf.to_csv(outfile,index=None)
```
